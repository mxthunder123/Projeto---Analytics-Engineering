```python
import sqlite3
import pandas as pd
from sqlalchemy import create_engine
```

```python
df_listing_2024_marco = pd.read_csv('listings_2024_março.csv')
df_listing_2024_marco["mes/ano"] = "marco/2024"
```

```python
df_listing_2024_junho = pd.read_csv('listings_junho_2024.csv')
df_listing_2024_junho["mes/ano"] = "junho/2024"
```

```python
df_listing_2023_dezembro = pd.read_csv('listings_dezembro_2023.csv')
df_listing_2023_dezembro["mes/ano"] = "dezembro/2023"
```

```python
df_listing_2023_setembro = pd.read_csv('listings_setembro_2023.csv')
df_listing_2023_setembro["mes/ano"] = "setembro/2023"
```

```python
df_listing_total = pd.concat([df_listing_2024_marco, df_listing_2024_junho, df_listing_2023_dezembro, df_listing_2023_setembro])
```

```python
df_listing_total
```

```python
df_reviews_2024_marco = pd.read_csv('reviews_2024_marco.csv',on_bad_lines="skip")
df_reviews_2024_marco["mes/ano"] = "marco/2024"
```

```python
df_reviews_2024_junho = pd.read_csv('reviews_junho_2024.csv')
df_reviews_2024_junho["mes/ano"] = "junho/2024"
```

```python
df_reviews_2023_dezembro = pd.read_csv('reviews_dezembro_2023.csv')
df_reviews_2023_dezembro["mes/ano"] = "dezembro/2023"
```

```python
df_reviews_2023_setembro = pd.read_csv('reviews_setembro_2023.csv')
df_reviews_2023_setembro["mes/ano"] = "setembro/2023"
```

```python
df_reviews_global = pd.concat([df_reviews_2024_marco, df_reviews_2024_junho, df_reviews_2023_dezembro, df_reviews_2023_setembro])
```

```python
conn = sqlite3.connect('airbnb_bronze.db')
df_listing_total.to_sql('listings_bronze', conn, if_exists='replace', index=False)
df_reviews_global.to_sql('reviews_bronze', conn, if_exists='replace', index=False)
#calendar.to_sql('calendar_bronze', conn, if_exists='replace', index=False)
```

```python

listings_clean = df_listing_total.drop_duplicates().rename(columns=lambda x: x.strip().lower().replace(" ", "_"))
reviews_clean = df_reviews_global.drop_duplicates().rename(columns=lambda x: x.strip().lower().replace(" ", "_"))
#calendar_clean = calendar.drop_duplicates().rename(columns=lambda x: x.strip().lower().replace(" ", "_"))


listings_clean['price'] = pd.to_numeric(listings_clean['price'].replace('[\$,]', '', regex=True), errors='coerce')
listings_clean['price'].fillna(listings_clean['price'].mean(), inplace=True)


listings_clean.to_sql('listings_silver', conn, if_exists='replace', index=False)
reviews_clean.to_sql('reviews_silver', conn, if_exists='replace', index=False)
#calendar_clean.to_sql('calendar_silver', conn, if_exists='replace', index=False)

conn.commit()
print("Dados limpos e armazenados na camada Silver.")
```

```python

assert listings_clean['id'].notnull().all(), "Existem IDs nulos nas listagens"
assert reviews_clean['listing_id'].notnull().all(), "Existem IDs de listagem nulos nas avaliações"


valid_ids = listings_clean['id'].unique()
assert reviews_clean['listing_id'].isin(valid_ids).all(), "Existem avaliações sem listagens correspondentes"

print("Validações de qualidade concluídas.")
```

```python
import pandera as pa
from pandera import Column, DataFrameSchema

# Definir o esquema de validação para os dados
# Definir o schema para validação
schema = DataFrameSchema({
    "neighbourhood": Column(pa.String, nullable=True),  # Permitindo valores nulos na coluna 'neighbourhood'
    "avg_price": Column(pa.Float, checks=[pa.Check.ge(100), pa.Check.le(2000)])  # Garantindo que 'avg_price' seja entre 100 e 2000
})

# Validar o DataFrame contra o esquema
validated_df = schema.validate(df.head(10))

# Exibir o DataFrame validado
print(validated_df.head())
```

```python
from pathlib import Path

# Criar o arquivo de configuração profiles.yml
profiles_dir = Path.home() / ".dbt"
profiles_dir.mkdir(parents=True, exist_ok=True)

with open(profiles_dir / "profiles.yml", "w") as f:
    f.write("""
airbnb_project:
  target: dev
  outputs:
    dev:
      type: sqlite
      threads: 1
      database: airbnb_bronze.db  # Caminho para o seu banco de dados SQLite
    """)
```

```python
# Criar os diretórios de modelos
models_dir = '/content/dbt_airbnb/airbnb_project/models'
os.makedirs(models_dir, exist_ok=True)

# Definir o modelo 'price_by_neighbourhood.sql'
price_by_neighbourhood_model = """
WITH avg_price AS (
    SELECT neighbourhood, AVG(price) AS avg_price
    FROM {{ ref('listings_bronze') }}  -- Refere-se à tabela 'listings_bronze' no seu banco
    GROUP BY neighbourhood
)
SELECT
    neighbourhood,
    avg_price
FROM avg_price
"""

# Escrever o modelo SQL no arquivo
with open(os.path.join(models_dir, 'price_by_neighbourhood.sql'), 'w') as f:
    f.write(price_by_neighbourhood_model)

# Definir outro modelo para a tabela listings_bronze
listings_bronze_model = """
SELECT * FROM {{ source('airbnb_data', 'listings_bronze') }}
"""

# Escrever o modelo SQL no arquivo
with open(os.path.join(models_dir, 'listings_bronze.sql'), 'w') as f:
    f.write(listings_bronze_model)

# Definir outro modelo para a tabela listings_silver
listings_silver_model = """
SELECT * FROM {{ source('airbnb_data', 'listings_silver') }}
"""

# Escrever o modelo SQL no arquivo
with open(os.path.join(models_dir, 'listings_silver.sql'), 'w') as f:
    f.write(listings_silver_model)

# Definir outro modelo para a tabela listings_gold
listings_gold_model = """
SELECT * FROM {{ source('airbnb_data', 'listings_gold') }}
"""

# Escrever o modelo SQL no arquivo
with open(os.path.join(models_dir, 'listings_gold.sql'), 'w') as f:
    f.write(listings_gold_model)

print("Modelos SQL criados com sucesso!")
```
